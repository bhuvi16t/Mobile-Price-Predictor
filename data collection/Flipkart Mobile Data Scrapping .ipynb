```python
final = pd.DataFrame()

model_name = []
price = []
memory = []
rating = []  # corrected spelling from 'ratting' to 'rating'
Display = []
Camera = []
Battery = []  # corrected spelling from 'Bettery' to 'Battery'
processor = []

# Loop through pages 1 to 59
for page in range(1,50):
    url = f'https://www.flipkart.com/search?q=smartphones+under+30000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={page}'
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36',
        'Accept-Language': 'en-US,en;q=0.5',
        'Referer': 'https://www.flipkart.com/',
        'Connection': 'keep-alive'
    }

    # Fetch the page
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, 'html.parser')

    # Find all the product containers on the page
    mobiles = soup.find_all('a','div', class_='CGtC98')  # Updated class name for product container

    # Loop through each mobile and extract details
    for mobile in mobiles:
        model = mobile.find('div', class_='KzDlHZ')  # Class for model name
        price_data = mobile.find('div', class_='yRaY8j ZYYwLA')  # Class for price
        rating_data = mobile.find('div', class_='XQDdHH')  # Class for rating
        
        # Append extracted data to lists
        model_name.append(model.text if model else None)
        price.append(price_data.text if price_data else None)
        rating.append(rating_data.text if rating_data else None)

        # Features such as memory, display, camera, etc.
        features = mobile.find_all('li', class_='J+igdf')
        memory.append(features[0].text if len(features) > 0 else None)
        Display.append(features[1].text if len(features) > 1 else None)
        Camera.append(features[2].text if len(features) > 2 else None)
        Battery.append(features[3].text if len(features) > 3 else None)
        processor.append(features[4].text if len(features) > 4 else None)
        data = {
                'Model Name': model_name,
                'Price': price,
                'Memory': memory,
                'Rating': rating,
                'Display': Display,
                'Camera': Camera,
                'Battery': Battery,
                'Processor': processor
            }

        df = pd.DataFrame(data)

    # Print progress for each page
    final = pd.concat([final, df], ignore_index=True)
    print(f"Page {page} data fetched.")
final    
# Create a DataFrame from the lists
```

```python
final.to_csv('Flipkart Mobile under 30k.csv')
```
